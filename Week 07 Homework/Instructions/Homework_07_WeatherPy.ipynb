```python
#Import Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import requests as req
from citipy import citipy
import random
import itertools
```

```python
#Read in City Data
world_cities = pd.read_csv("worldCities.csv")

city_names = world_cities
city_names = city_names.drop('Country', axis=1)
city_names = city_names.drop('Latitude', axis=1)
city_names = city_names.drop('Longitude', axis=1)
city_names = city_names.values.T.tolist()

city_names = list(itertools.chain(*city_names))
city_count = len(city_names)
city_count
```

```python
random_cities = []
n = 0
city_count = 500

while n <= city_count:
    try:
        if random_cities not in random_cities:
            random_cities.append(random.sample(city_names, 1))
            n = n + 1
    except:
        continue

random_cities = list(itertools.chain(*random_cities))
```

```python
# Get data for: (1) Temperature, (2) Humidity, (3) Cloudiness, (4) Wind Speed

# Save all of your "config" information within some variables. This includes your API key, the base URL, and the query terms desired.
api_key = "25bc90a1196e6f153eece0bc0b0fc9eb"
url = "http://api.openweathermap.org/data/2.5/weather?"
units = 'metric'

city_id = []
lat = []
temp = []
humidity = []
cloudiness = []
wind_speed = []

i = 1

for city in random_cities:
    
    try:
        query_url = url + "appid=" + api_key + "&q=" + city + "&units=" + units

        weather_response = req.get(query_url)
        weather_json = weather_response.json()

        print('City #: ' + str(i))
        print('Name: ' + city)
        print('URL: ' + query_url)
        print('-------------------------------------------------------------------------------------------------------------------')

        i = i + 1

        # Retrieve the current weather data from the JSON response.
        city_id.append(weather_json['id'])
        lat.append(weather_json['coord']['lat'])
        temp.append(weather_json['main']['temp'])
        humidity.append(weather_json['main']['humidity'])
        cloudiness.append(weather_json['clouds']['all'])
        wind_speed.append(weather_json['wind']['speed'])
    
    except:
        continue
```

```python
lat_lim = round(max(abs(min(lat)), abs(max(lat))) + 2, 0)
print(lat_lim)
```

```python
# Create Plots

# Plot 1: Latitude vs. Temperature
scatter_temp = plt.scatter(lat, temp, marker="o", s = 50, color = 'red', edgecolor = 'black', alpha = 0.75)

# Incorporate the other graph properties
plt.title("Latitude vs. Temperature")
plt.ylabel("Temperature (Celsius)")
plt.xlabel("Latitude")
plt.xlim((-lat_lim, lat_lim))
plt.grid(True, color = 'grey', alpha = 0.33, linewidth = 1)

# Save the figure
plt.savefig("Scatterplots/1_Lat_vs_Temp.png")

# Show plot
plt.show()
```

```python
# Create Plots

# Plot 2: Latitude vs. Humidity
scatter_humidity = plt.scatter(lat, humidity, marker="o", s = 50, color = 'gold', edgecolor = 'black', alpha = 0.75)

# Incorporate the other graph properties
plt.title("Latitude vs. Humidity")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.xlim((-lat_lim, lat_lim))
plt.grid(True, color = 'grey', alpha = 0.33, linewidth = 1)

# Save the figure
plt.savefig("Scatterplots/3_Lat_vs_Humidity.png")

# Show plot
plt.show()
```
